# The theory of information. Huffman's Algorithm #

*1.Считаем текст из файла*

```python
with open('text.txt') as f:
    s=f.read()
print(s)
```

*2. Создаем словарь, в котором считаем частоту встречи символов текста*

```python
d = {}
for i in s:
    if i in d:
        d[i]+=1
    else:
        d[i]=int(1)
print(d)
```

*3. Находим относительную частоту появления символов*  

```python
max_val = 100000
n=s[0]
v=d[n]
q = {}
while max_val!=0:
    max_val=max(d.values())
    for i,j in d.items():
        if j==max_val:
            q[i]=j/len(s)
            d[i]=0
            break
q[n]=v/len(s)
print(q)
```

*4. Сформируем и инициализируем словарь, в котором будем хранить коды символов.*

```python
kod = q.copy()
for i in kod:
    kod[i] = ''
print(kod)
```

*5. Алгоритм Хаффмана*

```python
w_sym = list(kod.keys())
w_nu = [q[i] for i in q]
for_l = {i:j for i,j in q.items()}
while len(w_sym)!=1:
    for i in w_sym[-1]:
        kod[i]+='0'
    for i in w_sym[-2]:
        kod[i]+='1'
    w_nu[-2]+=w_nu[-1]
    w_sym[-2]+=str(w_sym[-1])
    del w_nu[-1]
    del w_sym[-1]
    for i in range(1, len(w_nu)):
        if w_nu[-i]>w_nu[-i-1]:
            w_nu[-i], w_nu[-i-1] = w_nu[-i-1], w_nu[-i]
            w_sym[-i], w_sym[-i-1] = w_sym[-i-1], w_sym[-i]
for i in kod.keys():
    kod[i]=''.join(reversed(str(kod[i])))
print(kod)
```

*6. Запишем наш текст в закодированном виде во второй файл*

```python
wr = open('text2.txt', 'w')
for i in s:
    wr.write(kod[i])
wr.close()
with open('text2.txt') as f:
    sw=f.read()
print(sw)
```

*7. Раскодируем данное сообщение*

```python
decod={}
for i,j in kod.items():
    decod[j]=i
w=''
for i in sw:
    w+=str(i)
    if str(w) in decod:
        print(decod[w], end='')
        w=''
```

*8. Посчитаем среднюю длину слова, а также длину слова при равномерном кодировании*

```python
import math 
l = 0
for i in kod.keys():
    l+=for_l[i]*len(kod[i])
lconst=math.ceil(math.log2(len(kod)))
print('Средняя длина по Хаффману: ', l)
print('Средняя длина равномерного кода: ', lconst)
```

*9. Посчитаем энтропию и узнаем, является ли данное кодирование оптимальным*

```python
H=0
for i in for_l.keys():
    H+= (-for_l[i])*math.log2(for_l[i])
print('Энтропия равна: ',H)
print('Кодирование оптимальное' if H<=l<=(H+1) else 'Кодирование не оптимальное')
```
